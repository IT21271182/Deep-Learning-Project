```python
target_column = 'temperature_2m_mean (°C)'
features = df.drop(columns=['date', target_column])
```

### Creating Time Series Data

#### Create Sequences

#### Define a function to create sequences of data based on a given look-back window (e.g., the last 7 days to predict the next day’s temperature).

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense
```

```python
target = df[target_column].values

def create_sequences(features, target, look_back=7):
    X, y = [], []
    for i in range(len(features) - look_back):
        X.append(features[i:i + look_back])
        y.append(target[i + look_back])
    return np.array(X), np.array(y)

look_back = 7
X, y = create_sequences(features, target, look_back)
```

#### Split the data into training and testing sets for model evaluation.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

#### Define the GRU Model Architecture

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout

# Assuming your data preparation steps are done and you have X, y ready

# Confirm the input shape
print("Input shape:", X.shape)  # Should be (samples, look_back, features)


```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout, BatchNormalization

# Build the enhanced GRU model
model = Sequential()

# First GRU layer with Batch Normalization and Dropout
model.add(GRU(units=128, return_sequences=True, input_shape=(look_back, X_train.shape[2])))
model.add(BatchNormalization())
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting

# Second GRU layer
model.add(GRU(units=64, return_sequences=True))
model.add(BatchNormalization())
model.add(Dropout(0.2))

# Third GRU layer
model.add(GRU(units=32))
model.add(BatchNormalization())
model.add(Dropout(0.2))

# Dense layer to output the prediction
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Print model summary
model.summary()
```

```python
# Train the model with the training data
history = model.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.1)
```

```python
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss}, Test MAE: {test_mae}')
```

```python
# Predict on the test data
y_pred = model.predict(X_test)
y_pred.shape
```

```python
X_test.shape
```

```python
import numpy as np

# Create an array with the same number of features as the scaler expects
y_pred_reshaped = np.hstack((y_pred, np.zeros((y_pred.shape[0], scaler.n_features_in_ - 1))))

# Inverse transform the predictions
y_pred_actual = scaler.inverse_transform(y_pred_reshaped)[:, 0]  # Select the first feature
y_test_actual = scaler.inverse_transform(np.hstack((y_test.reshape(-1, 1), np.zeros((y_test.shape[0], scaler.n_features_in_ - 1)))))[:, 0]
```

```python
# Displaying the predictions and original values

# Create a DataFrame for better visualization
predictions_df = pd.DataFrame({
    'Actual Temperature (°C)': y_test_actual.flatten(),
    'Predicted Temperature (°C)': y_pred_actual.flatten()
})

print(predictions_df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(y_test_actual, label='Actual Temperature')
plt.plot(y_pred_actual, label='Predicted Temperature')
plt.title('Actual vs Predicted Temperature')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Calculate MSE
mse = mean_squared_error(y_test_actual, y_pred_actual)
# Calculate RMSE
rmse = np.sqrt(mse)
# Calculate R²
r_squared = r2_score(y_test_actual, y_pred_actual)

print(f'Mean Squared Error (MSE): {mse:.4f}')
print(f'Root Mean Squared Error (RMSE): {rmse:.4f}')
print(f'R-squared (R²): {r_squared:.4f}')
```

Visualize Residuals (the differences between actual and predicted values)



```python
# Calculate residuals
residuals = y_test_actual - y_pred_actual

# Plot residuals
plt.figure(figsize=(12, 6))
plt.scatter(range(len(residuals)), residuals, color='purple', alpha=0.5)
plt.axhline(0, color='red', linestyle='--', linewidth=1)
plt.title('Residuals Plot')
plt.xlabel('Time Steps')
plt.ylabel('Residuals (Actual - Predicted)')
plt.grid()
plt.show()
```
