##### Data Preprocessing
Load the Dataset

```python
#Importing Libraries needed 
import pandas as pd
```

```python
# Load the weather data
weather_data = pd.read_csv('DataSet\weatherData.csv')

# Load the location data
location_data = pd.read_csv('DataSet\locationData.csv')
```

```python
# Display the first few rows of weather data
print("Weather Data:")
weather_data.head()
```

```python
# Display the first few rows of location data
print("Location Data:")
location_data.head()
```

###### Merge data frames

```python
# Merge the dataframes on 'location_id'
merged_data = pd.merge(weather_data, location_data, on='location_id')  
```

```python
# Display the first few rows of the merged dataframe
print("\nMerged Data:")
merged_data.head()
```

Remove Unnecessary Columns

```python
# Define the columns to drop
columns_to_drop = [
    'weather_code (wmo code)',
    'daylight_duration (s)',
    'sunshine_duration (s)',
    'latitude',
    'longitude',
    'elevation',
    'utc_offset_seconds',
    'timezone',
    'timezone_abbreviation',
    'sunrise',
    'sunset',
]

# Remove unnecessary columns
merged_data = merged_data.drop(columns=columns_to_drop)

# Display the first few rows of the modified merged data
print("\nMerged Data after dropping unnecessary columns:")
merged_data.head()
```

Handle Missing Values

```python
# Check for missing values in the merged_data
print("Missing values in each column before handling:")
print(merged_data.isnull().sum())
```

One-Hot Encoding

```python
# Perform one-hot encoding on the 'city_name' column
merged_data = pd.get_dummies(merged_data, columns=['city_name'], drop_first=True)

# Display the first few rows of the updated merged_data to check the result
merged_data.head()
```

Handling the date

```python
# Convert the date column to datetime format
merged_data['date'] = pd.to_datetime(merged_data['date'])
```

```python
# Verify the data types of all columns in the merged_data DataFrame
print(merged_data.dtypes)
```

```python
import matplotlib.pyplot as plt

# List all the one-hot encoded city columns
city_columns = [col for col in merged_data.columns if 'city_name_' in col]

# Number of cities
num_cities = len(city_columns)

# Set the number of rows and columns for subplots (adjust as needed)
n_cols = 3  # You can change the layout to have more or fewer columns
n_rows = (num_cities // n_cols) + (num_cities % n_cols > 0)

# Create a figure for subplots
fig, axs = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

# Flatten the axs array for easy iteration
axs = axs.flatten()

# Plot for each city
for i, city in enumerate(city_columns):
    # Filter the data for the current city
    city_data = merged_data[merged_data[city] == 1]
    
    # Plot the temperature data for the current city
    axs[i].plot(
        city_data['date'],  # Filtered date
        city_data['temperature_2m_mean (°C)'],  # Filtered temperature data
        label=city.split('_')[-1],  # Simplified city name
        marker='o',
        linestyle='-'
    )
    
    # Set the title and labels for each subplot
    axs[i].set_title(f"Temperature Trend: {city.split('_')[-1]}")
    axs[i].set_xlabel('Date')
    axs[i].set_ylabel('Mean Temperature (°C)')
    axs[i].tick_params(axis='x', rotation=45)  # Rotate x-axis labels

# Remove any unused subplots
for j in range(i + 1, len(axs)):
    fig.delaxes(axs[j])

# Show the plot
plt.tight_layout()
plt.show()
```

##### Prepare Data for LSTM
Feature Selection


```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
# Define the features (columns that want to use for prediction) and target variable (temperature)
features = [
    'temperature_2m_min (°C)', 
    'temperature_2m_max (°C)',
    'apparent_temperature_min (°C)', 
    'apparent_temperature_max (°C)',
    'precipitation_sum (mm)',
    'rain_sum (mm)'
]

# Add the one-hot encoded city columns to the feature list
city_columns = [col for col in merged_data.columns if 'city_name_' in col]
features.extend(city_columns)

# Target variable
target = 'temperature_2m_mean (°C)'
```

Scale Features

```python
scaler = MinMaxScaler()

# Apply the scaler to the features and target separately
scaled_features = scaler.fit_transform(merged_data[features])
scaled_target = scaler.fit_transform(merged_data[[target]])

# Convert the scaled data back into DataFrames for easier handling
scaled_features_df = pd.DataFrame(scaled_features, columns=features)
scaled_target_df = pd.DataFrame(scaled_target, columns=[target])

# Combine the scaled features and target back into one DataFrame
scaled_data = pd.concat([scaled_features_df, scaled_target_df], axis=1)
```

```python
scaled_data
```

```python
for col in scaled_data:
    print(f'{col}: {scaled_data[col].unique()}')
```

Split the Data

```python
# Split the data without shuffling to maintain the time-series structure
train_size = int(len(scaled_data) * 0.8)  # Use 80% for training, 20% for testing

train_data = scaled_data.iloc[:train_size]
test_data = scaled_data.iloc[train_size:]

# Separate features and target for training and testing
X_train = train_data[features].values
y_train = train_data[target].values

X_test = test_data[features].values
y_test = test_data[target].values

# Check the shapes of the resulting arrays
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

##### Reshape Data for LSTM

```python
import numpy as np
```

```python
# Define the number of time steps (e.g., how many previous days to consider for predicting the next day)
n_time_steps = 10  # You can adjust this value based on how far back you want to look

# Prepare features and target (temperature to predict)
# Features: You can select other features as well, depending on your dataset
features = ['temperature_2m_max (°C)', 'temperature_2m_min (°C)', 'temperature_2m_mean (°C)', 
            'apparent_temperature_max (°C)', 'apparent_temperature_min (°C)', 'apparent_temperature_mean (°C)', 
            'precipitation_sum (mm)', 'rain_sum (mm)'] + city_columns

target = 'temperature_2m_mean (°C)'
```

```python
# Convert merged_data to a NumPy array for faster access
data_array = merged_data[features + [target]].values

# Initialize lists to store sequences
X = []
y = []

# Loop through the dataset and create sequences of length n_time_steps
for i in range(n_time_steps, len(data_array)):
    X.append(data_array[i - n_time_steps:i, :-1])  # Previous 'n_time_steps' features (exclude target)
    y.append(data_array[i, -1])  # Target temperature for the next day

# Convert the lists to numpy arrays for LSTM input
X = np.array(X)
y = np.array(y)

print(f"Shape of X: {X.shape}")  # Should be (samples, time steps, features)
print(f"Shape of y: {y.shape}")  # Should be (samples,)
```

##### Build and Train the LSTM Model
Define the LSTM Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Step 1: Check shapes
print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
```

```python
# Reshape the data if necessary
# Check if X_train has only 2 dimensions
if len(X_train.shape) == 2:
    X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
if len(X_test.shape) == 2:
    X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])

print("New shape of X_train:", X_train.shape)
print("New shape of X_test:", X_test.shape)
```

```python
# Define the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))  # Dropout layer to prevent overfitting
model.add(LSTM(50, return_sequences=False))  # Second LSTM layer
model.add(Dropout(0.2))
model.add(Dense(25))  # Dense layer for learning nonlinear combinations
model.add(Dense(1))  # Output layer for regression task
```

Compile the Model 

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

Fit and train the Model

```python
# Set the number of epochs and batch size
epochs = 50  
batch_size = 32  

# Train the model
history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, 
                    validation_data=(X_test, y_test), verbose=1)
```

Plot the training history

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
